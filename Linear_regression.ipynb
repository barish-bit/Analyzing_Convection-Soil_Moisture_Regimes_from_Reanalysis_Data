```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
from sklearn.metrics import r2_score

data = pd.read_csv('/content/Pune_soil_rain_2020.csv')
list=[0,50,100,150,200,250,300,350]

for i in range (7):
  rain_data_list=data["tp"][list[i]:list[i+1]].tolist()
  soil_data_list=data["src"][int(list[i])+10:int(list[i+1])+10].tolist()

  plt.scatter(rain_data_list, soil_data_list)
  soil_data_list= np.array(soil_data_list)
  rain_data_list= np.array(rain_data_list)
  #print(soil_data_list)
  a, b = np.polyfit(rain_data_list, soil_data_list, 1)
  r2 = r2_score(soil_data_list, a*rain_data_list+b)
  plt.plot(rain_data_list, a*rain_data_list+b, color="red", label="r2_value"+str(r2))

  plt.title('Rainfall_Soil_Linear_Regression_Pune')
  plt.xlabel('RAINFALL_Days_from_'+str(list[i])+"_to_"+str(list[i+1]))
  plt.ylabel('Soil_water')
  plt.legend(loc="upper left")
  #print('r2 score for perfect model is', r2)


  plt.show()
```
